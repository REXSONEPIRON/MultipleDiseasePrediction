```python
from datetime import datetime, timedelta
```

```python
def convert_days_to_exact_age(days):
    today = datetime.today()
    birth_date = today - timedelta(days=int(days))

    age_years = today.year - birth_date.year
    age_months = today.month - birth_date.month
    age_days = today.day - birth_date.day

    # Adjust for negative values
    if age_days < 0:
        age_months -= 1
        prev_month_last_day = (birth_date.replace(month=birth_date.month % 12 + 1, day=1) - timedelta(days=1)).day
        age_days += prev_month_last_day

    if age_months < 0:
        age_years -= 1
        age_months += 12

    return f"{age_years}"
```

```python
df["Exact_Age"] = df["age"].apply(convert_days_to_exact_age)
```
